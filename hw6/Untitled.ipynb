```python
import MLutil
import matplotlib.pyplot as plt
import numpy as np
from collections import OrderedDict
```

```python
storer = MLutil.Storer('mf')
```

```python
storer.show_key()
```

```python
storer.show_record_item()
```

```python
np.min(storer.dict['model100']['val_rmse'])
```

```python
dim_list = []
dim_train = []
for dim in range(10,301,10):
    dim_list.append(np.min(storer.dict['model' + str(dim)]['val_rmse']))
    dim_train.append(np.min(storer.dict['model' + str(dim)]['rmse']))
```

```python
dim_list
```

```python
plt.plot(range(10, 301, 10), dim_list, label = 'validation')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.title('Laten Dimenseion Difference')
plt.ylabel('valication rmse')
plt.xlabel('dimenseion')
plt.show()

print(np.argmin(dim_list))

plt.close()
```

```python
nobias = storer.dict['model_nobias_250']['val_rmse']
bias = storer.dict['model240']['val_rmse']
plt.plot(nobias, label = 'no bias')
plt.plot(bias, label = 'bias')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.title('Bias vs. No bias')
plt.xlabel('epoch')
plt.ylabel('rmse')
plt.show()
plt.close()
```

```python
bias = storer.dict['model240']['val_rmse'][3:]
dnn = storer.dict['model10']['val_rmse'][3:len(bias) + 3]
plt.plot(range(3,len(bias)+3), bias, label = 'mf')
plt.plot(range(3,len(bias)+3), dnn, label = 'dnn')
plt.title('dnn vs. mf')
plt.xlabel('epoch')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
plt.close()
```
