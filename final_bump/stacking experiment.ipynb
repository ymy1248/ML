## 對比dt 和 rf 兩個不同stacking方法

```python
stack_rf_score = pickle.load(open('stack_rf_score.pkl', 'rb'))
stack_avg_score = pickle.load(open('stack_avg_score.pkl', 'rb'))
x_axis = np.arange(1, 20, 1)
x_axis = 1 - 0.015 * x_axis 
plt.grid(True)
plt.plot(x_axis, stack_rf_score, label = 'rf voter')
plt.plot(x_axis, stack_avg_score, label = 'average')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.title('compare average voter with rf voter in weaker models')
plt.xlabel('classifier training data ratio')
plt.ylabel('accuracy')
plt.show()
```

```python
plt.plot(x_axis, stack_rf_score, label = 'svc + rf')
plt.plot(x_axis, rf_voter_score, label = 'xgb + rf')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.grid(True)
plt.title('compare different models stacking')
plt.xlabel('model training data ratio')
plt.ylabel('accuracy')
plt.show()
```
