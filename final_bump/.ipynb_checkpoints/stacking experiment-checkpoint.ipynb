# Experiment

```python
%matplotlib inline
import pickle
import matplotlib.pyplot as plt
import numpy as np
```

## score plot

```python
svc_voter_score = pickle.load(open('svc_voter_score.pkl', 'rb'))
rf_voter_score = pickle.load(open('svm_voter_score.pkl', 'rb'))
xgb_voter_score = pickle.load(open('xgb_voter_score.pkl', 'rb'))

pickle.dump(rf_voter_score, open('rf_voter_score.pkl', 'wb'))
```

```python
svc_voter_score = np.array(svc_voter_score)
rf_voter_score = np.array(rf_voter_score)
average_voter_score = svc_voter_score + rf_voter_score
average_voter_score /= 2 2 
plt.plot(average_voter_score)

x_axis = np.arange(1, 20, 1)
x_axis = 1 - 0.015 * x_axis 

plt.plot(x_axis, average_voter_score, label = 'average')
plt.plot(x_axis, rf_voter_score, label = 'random forest')
plt.plot(x_axis, svc_voter_score, label = 'svc')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.title('model training data partition result')
plt.xlabel('model training data partition')
plt.ylabel('accuracy')
plt.show()
plt.close()
```
