```python
import pickle
from collections import OrderedDict

score = pickle.load(open("model_score", "rb"))
```

```python
score = OrderedDict(sorted(score.items()))
pickle.dump(score, open("model_score", "wb"))

for key in score.keys():
    print(key)
delete = []
for key, value in score.items():
    if len(value["vali"]) == 0:
        delete.append(key)
for k in delete:
    score.pop(k)
```

```python
for key, value in score.items():
    print(key)
    print("validation:", max(value.get("vali", [0,0])))
    print("train:", max(value["train"]))
    print("vali_loss:", min(value.get("vali_loss", [0])))
```
